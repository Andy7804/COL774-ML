**Logistic Regression using Newton's Method**

**Setup and Preliminaries**
<br><br>
(1) Import the necessary libraries <br>
(2) Get the dataset from the CSV files

```python
# (1) Import the necessary libraries
import numpy as np
import pandas as pd
from pathlib import Path

# (2) Get the dataset from the CSV files (relative to the ASSIGNMENT1 folder)
# Get the absolute path of the ASSIGNMENT1 folder (i.e. the parent folder)
parent_folder = Path.cwd().parent  # Moves up from Q4 to ASSIGNMENT1

# Define the data folder relative to ASSIGNMENT1
data_folder = parent_folder / "data" / "Q3"

# Define file paths for q4x.dat and q4y.dat (case-sensitive filenames)
x_file = data_folder / "logisticX.csv"
y_file = data_folder / "logisticY.csv"

# Load the datasets - CSV to Pandas df to Numpy array
X = pd.read_csv(x_file, header=None).values  
y = pd.read_csv(y_file, header=None).values.ravel()

# Display the shape (verification purposes)
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

(3) Split the dataset into 'train' and 'test' datasets

```python
# Set random seed for reproducibility
np.random.seed(42)

# Get the number of samples
num_samples = X.shape[0]

# Shuffle the indices
shuffled_indices = np.random.permutation(num_samples)

# Define the split index (85% train, 15% test)
split_idx = int(num_samples * 0.80)

# Split the indices
train_indices = shuffled_indices[:split_idx]
test_indices = shuffled_indices[split_idx:]

# Create train and test sets
X_train, y_train = X[train_indices], y[train_indices]
X_test, y_test = X[test_indices], y[test_indices]

# Display the shapes for verification
print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

**1 &emsp; Train the logistic regresion model (on train data)**

```python
from logistic_regression import LogisticRegressor
# Instantiate and train the logistic regression model
model = LogisticRegressor()
model.fit(X_train, y_train)
print(model.theta)
```

```python
from logistic_regression import plot_decision_boundary
plt = plot_decision_boundary(model, X, y)
plt.show()
```
